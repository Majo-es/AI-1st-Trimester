```python
# Importamos las librerías necesarias
import pandas as pd

# Cargamos el archivo csv en un DataFrame
df = pd.read_csv("/content/drive/MyDrive/bank_numeric.csv", index_col=0, sep="," ) # Unnamed is an index.

# Mostramos el tamaño del df
print("Tamaño df:\n", df.shape)

# Mostramos el nombre de las columnas
print("Columnas del df:\n", df.columns)

# Visualizamos
df.head()
```

```python
df.info()
# ALL ARE NUMERICAL
# NO MISSING VALUES
```

```python
# IDENTIFY THE X(FEATURES) AND y (objective variable):
y = df.y_yes
X = df.drop("y_yes", axis = 1)
```

```python
# DIVIDE DATA IN TRAIN & TEST
# USE SCIKIT-LEARN
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, train_size=0.8, random_state=42)
```

```python
# DEFINE THE MODEL TO USE & TRAIN IT
# DECISION TREE
from sklearn.tree import DecisionTreeClassifier

# CALL IT
model = DecisionTreeClassifier(max_depth=3, class_weight="balanced")
# max_depth=3 PARAMETER TO TRIM THE TREE
# class_weight="balanced" ASSIGNS WEIGHTS TO THE CLASSES IN THE DATA. USEFUL WHEN DATASET IS IMBALANCED!!

# TRAIN IT
model.fit(X_train, y_train)
# TO TRAIN ONLY X_test NOT y_test!!
predictions = model.predict(X_test)
```

```python
# CHECK ACCURACY
from sklearn.metrics import accuracy_score
print("Accuracy= ", accuracy_score(y_test, predictions))
```

```python
# TO SHOW METRICS
from sklearn.metrics import classification_report
print(classification_report(y_test, predictions))
# PRECISION ALMOST THE SAME AS ACCURACY
```

NOT WORKING. THERE ARE MORE CLIENTS THAT DO NOT ACCEPT THE CAMPAIGN THAN CLIENTS THAT DO.

```python
df.y_yes.value_counts()
```

IT'S UNBALANCED! TO FIX IT, HAVE MORE REAL DATA FROM CLIENTS THAT ACCEPT THE CAMPAIGN, CAMPAIGN MORE,ETC.

- DOWNSAMPLING (undersampling):reducing the number of samples in the majority class to match the minority class.

- UPSAMPLING (oversampling): increasing the number of samples in the minority class to match the majority class.

# RANDOM FOREST:

```python
from sklearn.ensemble import RandomForestClassifier
forest = RandomForestClassifier()
forest.fit(X_train, y_train)
```

```python
predictions_forest = forest.predict(X_test)
```

```python
print(accuracy_score(y_test, predictions_forest))
print(classification_report(y_test, predictions_forest))
```

## SOLVING UNBALANCE: DOWNSAMPLING WITH `SMOTE`

```python
from imblearn.over_sampling import SMOTE
smote = SMOTE(random_state=0)
X_train_balanced, y_train_balanced = smote.fit_resample(X_train, y_train)
```

```python
forest_balanced = RandomForestClassifier()
forest_balanced.fit(X_train_balanced, y_train_balanced)
```

```python
predictions_balanced = forest_balanced.predict(X_test)
print(classification_report(y_test, predictions_balanced))
```
